## **1. Bussiness Problem Understanding**

**_Context_**

Customer Lifetime Value (CLV) adalah metrik yang digunakan untuk memperkirakan total nilai pendapatan yang akan dihasilkan oleh seorang pelanggan selama masa hubungan mereka dengan sebuah perusahaan. CLV membantu perusahaan memahami nilai ekonomi jangka panjang dari pelanggan dan menginformasikan keputusan bisnis seperti alokasi anggaran pemasaran, strategi retensi pelanggan, dan penawaran khusus. Dengan menganalisis CLV, perusahaan dapat mengidentifikasi pelanggan yang paling berharga dan fokus pada strategi untuk mempertahankan mereka serta meningkatkan loyalitas, yang pada akhirnya dapat meningkatkan keuntungan dan pertumbuhan jangka panjang perusahaan.

Dataset ini berisi informasi terkait nilai seumur hidup pelanggan dalam **konteks asuransi kendaraan**. Data ini mencakup berbagai atribut seperti kelas kendaraan, jenis cakupan asuransi, tipe penawaran pembaruan, status pekerjaan, status pernikahan, dan tingkat pendidikan pelanggan. Selain itu, data ini juga mencatat jumlah polis, premi bulanan untuk asuransi kendaraan, jumlah klaim total, pendapatan, dan nilai seumur hidup pelanggan.

Beberapa contoh data menunjukkan pelanggan dengan berbagai profil, misalnya seorang pelanggan yang sudah pensiun dengan kendaraan empat pintu dan cakupan asuransi diperpanjang memiliki premi bulanan sebesar $90 dan total klaim sebesar $571.479. Data ini sangat berguna untuk analisis lebih lanjut dalam memahami faktor-faktor yang mempengaruhi nilai seumur hidup pelanggan dan bagaimana perusahaan asuransi dapat mengoptimalkan strategi pemasaran dan penawaran mereka.


**_Problem Statement_**

Salah satu tantangan utama yang dihadapi perusahaan asuransi adalah meningkatkan Customer Lifetime Value (CLV), yaitu total jumlah uang yang diharapkan akan dibelanjakan oleh pelanggan dengan bisnis Anda selama masa hubungan bisnis rata-rata. 
Salah satu masalah utama yang dihadapi perusahaan asuransi adalah rendahnya tingkat retensi pelanggan yang berdampak negatif pada Customer Lifetime Value (CLV), **yaitu total jumlah uang yang diharapkan akan dibelanjakan oleh pelanggan dengan bisnis atau produk yang ditawarkan**, selama masa hubungan bisnis rata-rata.

**_Goals_**

Salah satu tujuan utama perusahaan asuransi adalah meningkatkan Customer Lifetime Value (CLV), yaitu total jumlah uang yang diharapkan akan dibelanjakan oleh pelanggan dengan bisnis Anda selama masa hubungan bisnis rata-rata. Agar dapat mencapai hal ini, perusahaan harus memahami kebutuhan pelanggan secara mendalam dan menawarkan solusi yang lebih personal serta relevan. Untuk mendukung upaya ini, perusahaan perlu mengembangkan model prediksi yang mampu mengidentifikasi pelanggan dengan potensi CLV tinggi dan rendah, memungkinkan penerapan strategi retensi dan pemasaran yang lebih efektif dan tepat sasaran.

**Analytic Approach**

Jadi, yang perlu kita lakukan adalah menganalisis data untuk menemukan pola dari fitur-fitur yang ada, yang memengaruhi Customer Lifetime Value (CLV) pelanggan. Selanjutnya, kita akan membangun suatu model prediksi yang akan membantu perusahaan untuk menyediakan 'tool' prediksi CLV pelanggan baru, yang mana akan berguna untuk tim pemasaran dan penjualan dalam merancang strategi retensi dan penawaran yang lebih efektif.

**Metric Evaluation**

Untuk mengevaluasi kinerja model regresi prediksi CLV, kita akan menggunakan beberapa metrik evaluasi. Pertama, Mean Absolute Error (MAE) yang mengukur rata-rata kesalahan absolut antara nilai prediksi dan nilai aktual, memberikan gambaran langsung tentang seberapa akurat model kita. Kedua, Root Mean Squared Error (RMSE) yang memberikan penalti lebih besar untuk kesalahan yang lebih besar, sangat berguna untuk mengidentifikasi model yang lebih sensitif terhadap kesalahan besar, dan Mean Absolute Percentage Error (MAPE) yang mengukur kesalahan prediksi dalam bentuk persentase, memberikan perspektif yang lebih relatif dan mudah dipahami mengenai seberapa jauh prediksi model dari nilai aktual. Dengan menggunakan metrik-metrik ini, kita dapat secara efektif mengevaluasi dan meningkatkan model prediksi CLV kita.

## **2. Data Understanding**

Dataset ini berisi informasi yang berkaitan dengan nilai seumur hidup pelanggan (Customer Lifetime Value atau CLV) dalam konteks asuransi kendaraan. Data ini mencakup berbagai atribut yang berpotensi mempengaruhi CLV, antara lain:

1. Vehicle Class: Jenis kendaraan yang dimiliki oleh pelanggan, seperti Four-Door Car dan Two-Door Car.
2. Coverage: Jenis cakupan asuransi yang dipilih oleh pelanggan, seperti Basic, Extended, dan Premium.
3. Renew Offer Type: Jenis penawaran pembaruan polis asuransi yang diberikan kepada pelanggan, seperti Offer1, Offer2, dan lainnya.
4. Employment Status: Status pekerjaan pelanggan, misalnya Employed, Retired, Medical Leave, dll.
5. Marital Status: Status pernikahan pelanggan, seperti Single, Married, dan Divorced.
6. Education: Tingkat pendidikan pelanggan, seperti High School or Below, College, dan Master.
7. Number of Policies: Jumlah polis asuransi yang dimiliki oleh pelanggan.
8. Monthly Premium Auto: Jumlah premi bulanan yang dibayarkan oleh pelanggan untuk asuransi kendaraan.
9. Total Claim Amount: Total jumlah klaim yang diajukan oleh pelanggan.
10. Income: Pendapatan tahunan pelanggan.
11. Customer Lifetime Value: Nilai seumur hidup pelanggan, yaitu total jumlah uang yang diharapkan akan dibelanjakan oleh pelanggan dengan bisnis selama masa hubungan bisnis rata-rata.

### **5. Data Correlation**

**Fitur Kategorik**

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# List of categorical features including 'Number of Policies'
categorical_features = df_model.select_dtypes(include=['object']).columns.tolist()

# Perform ANOVA for each categorical feature
anova_results = []
for feature in categorical_features:
    model = ols(f'Q("Customer Lifetime Value") ~ C(Q("{feature}"))', data=df_model).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    anova_results.append({
        'Categorical Feature': feature,
        'ANOVA Statistic': anova_table['F'][0],
        'p-value': anova_table['PR(>F)'][0]
    })

# Convert results to DataFrame
anova_results_df = pd.DataFrame(anova_results)

anova_results_df
```

1. Fitur kategorikal seperti Vehicle Class, Coverage, Renew Offer Type, dan Number of Policies memiliki pengaruh yang signifikan terhadap CLV.
2. Marital Status dengan pval 0.750764899 akan di drop
3. Education memiliki pengaruh yang signifikan, tetapi tidak sebesar beberapa fitur lainnya.

Model yang dihasilkan nantinya hanya akan mampu memprediksi nilai-nilai yang berada dalam rentang data dari 'clean dataset' ini. Prediksi di luar rentang tersebut akan cenderung bias.


```python
# Membuat fungsi untuk mengecek outliers
def find_anomalies(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'limit_bawah: {limit_bawah}')
    print(f'limit_atas: {limit_atas}')
```

### **9. Encoding**

Encoding adalah langkah penting dalam preprocessing data yang memastikan bahwa data kategorikal dapat digunakan oleh algoritma machine learning, meningkatkan kinerja model, dan memastikan pemrosesan data yang konsisten dan adil. Proses ini mengubah kolom kategorikal menjadi numerik, sehingga memungkinkan algoritma untuk memahami dan memproses data tersebut dengan lebih efektif. Yang nantinya dapat diaplikasikan ke Pipeline Model

+ One Hot Encoder :
    - Vehicle_Class, Renew_Offer_Type : karenna memiliki jenis kategori yang bersifat nominal
+ Ordinal Encoder :
    - Educational : Karena memilliki tingkatan dan tidak bisa dirubah dari High School sampai Doctor
    - Coverage : Karena memiliki tingkatan, dan bukan ordinal dari Basic, Extended, dan Premium
+ Standard Scaler :
    - Akan dilakukan pada semua features kecuali target yang memiliki tipe data numerical

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler

# Define the categorical, ordinal, and numerical columns
categorical_features = ['Vehicle_Class', 'Renew_Offer_Type']
ordinal_features = ['Education', 'Coverage']
numerical_features = ['Monthly_Premium_Auto', 'Total_Claim_Amount', 'Income', 'Number_of_Policies']

# Define ordinal categories for Education and Coverage
education_categories = ['High School or Below', 'College', 'Bachelor', 'Master', 'Doctor']
coverage_categories = ['Basic', 'Extended', 'Premium']

# Create a ColumnTransformer with One-Hot Encoding, Ordinal Encoding, and Standard Scaling
preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features),
        ('ordinal_education', OrdinalEncoder(categories=[education_categories]), ['Education']),
        ('ordinal_coverage', OrdinalEncoder(categories=[coverage_categories]), ['Coverage']),
        ('num', StandardScaler(), numerical_features)
    ],
    remainder='passthrough'  # Pass through other columns unchanged
)

preprocessor
```

### **4.1 Model Benchmarking dan Cross Validation**

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, BaggingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import make_scorer, mean_squared_error, mean_absolute_error
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
import numpy as np
import pandas as pd

# Define custom MAPE scorer
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape_scorer = make_scorer(mean_absolute_percentage_error, greater_is_better=False)

# Identify numeric and categorical columns
numeric_features = xtrain.select_dtypes(include=['int64', 'float64']).columns
categorical_features = xtrain.select_dtypes(include=['object']).columns


# Create model pipelines
models = [
    ('LinearRegression', Pipeline([('preprocessor', preprocessor), ('regressor', LinearRegression())])),
    ('DecisionTree', Pipeline([('preprocessor', preprocessor), ('regressor', DecisionTreeRegressor(random_state=42))])),
    ('SVR', Pipeline([('preprocessor', preprocessor), ('regressor', SVR())])),
    ('RandomForest', Pipeline([('preprocessor', preprocessor), ('regressor', RandomForestRegressor(random_state=42))])),
    ('XGBoost', Pipeline([('preprocessor', preprocessor), ('regressor', XGBRegressor(random_state=42))])),
    ('Bagging', Pipeline([('preprocessor', preprocessor), ('regressor', BaggingRegressor(estimator=LinearRegression(), n_estimators=10, random_state=42))]))
]

# Set up KFold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=1)

# Perform cross-validation and store results
results = []

for name, model in models:
    # RMSE
    rmse_scores = np.sqrt(-cross_val_score(model, xtrain, ytrain, scoring='neg_mean_squared_error', cv=kf))
    
    # MAE
    mae_scores = -cross_val_score(model, xtrain, ytrain, scoring='neg_mean_absolute_error', cv=kf)
    
    # MAPE
    mape_scores = -cross_val_score(model, xtrain, ytrain, scoring=mape_scorer, cv=kf)
    
    results.append({
        'Model': name,
        'RMSE': rmse_scores.mean(),
        'MAE': mae_scores.mean(),
        'MAPE': mape_scores.mean()
    })

# Create DataFrame
results_df = pd.DataFrame(results)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

# Function to perform cross-validation and return results
def perform_cv(model, X, y, cv=5):
    rmse_scores = np.sqrt(-cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=cv))
    mae_scores = -cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv)
    mape_scores = -cross_val_score(model, X, y, scoring=mape_scorer, cv=cv)
    
    return {
        'RMSE': rmse_scores.mean(),
        'MAE': mae_scores.mean(),
        'MAPE': mape_scores.mean()
    }

# Perform cross-validation for RandomForest
rf_cv_results = perform_cv(grid_search_rf.best_estimator_, xtrain, ytrain)

# Perform cross-validation for DecisionTree
dt_cv_results = perform_cv(grid_search_dt.best_estimator_, xtrain, ytrain)

# Perform cross-validation for XGBoost
xgb_cv_results = perform_cv(grid_search_xgb.best_estimator_, xtrain, ytrain)

# Create a DataFrame to display results
cv_results_df = pd.DataFrame({
    'Model': ['RandomForest', 'DecisionTree', 'XGBoost'],
    'RMSE': [rf_cv_results['RMSE'], dt_cv_results['RMSE'], xgb_cv_results['RMSE']],
    'MAE': [rf_cv_results['MAE'], dt_cv_results['MAE'], xgb_cv_results['MAE']],
    'MAPE': [rf_cv_results['MAPE'], dt_cv_results['MAPE'], xgb_cv_results['MAPE']]
})

# Display results
print("Cross-validation results after hyperparameter tuning:")
print(cv_results_df)
```

### **5. Cara Kerja Model Terpilih**

### **RANDOM FOREST**

Random Forest adalah salah satu algoritma pembelajaran mesin yang populer untuk tugas regresi dan klasifikasi, termasuk kasus prediksi Customer Lifetime Value (CLV). Algoritma ini bekerja dengan membuat banyak pohon keputusan (decision trees) dan menggabungkan hasilnya untuk mendapatkan prediksi yang lebih akurat dan stabil.\
Berikut adalah cara kerja Random Forest pada kasus ini:

### **Cara Kerja Random Forest**

**1. Pembentukan Banyak Pohon Keputusan (Decision Trees):**

+ Random Forest membuat banyak pohon keputusan selama pelatihan. Setiap pohon dibangun menggunakan subset acak dari data pelatihan dan subset acak dari fitur. Ini dikenal sebagai teknik bagging atau Bootstrap Aggregating.
+ Karena setiap pohon dilatih pada sampel data dan fitur yang berbeda, mereka cenderung menghasilkan variasi dalam prediksi mereka, yang membantu dalam menangkap pola yang kompleks dalam data.

**2. Pemilihan Fitur secara Acak:**
+ Pada setiap node di pohon, hanya sebagian kecil dari semua fitur yang dipertimbangkan untuk split. Ini membantu mengurangi korelasi antara pohon dan meningkatkan keanekaragaman model.
+ Dalam konteks prediksi CLV, ini berarti bahwa Random Forest dapat mempertimbangkan berbagai kombinasi faktor seperti pendapatan, jenis kendaraan, dan premi bulanan untuk membuat keputusan.

**3. Pembentukan Keputusan dari Setiap Pohon:**
+ Setiap pohon dalam hutan memberikan prediksi terpisah untuk setiap observasi. Untuk tugas regresi, prediksi akhir dari Random Forest adalah rata-rata dari semua prediksi pohon individu.

**4. Penggabungan Hasil (Averaging):**
+ Dengan menggabungkan hasil dari semua pohon, Random Forest dapat mengurangi varians prediksi dan menangani overfitting lebih baik dibandingkan dengan satu pohon keputusan.
+ Proses ini menghasilkan prediksi yang **lebih stabil dan akurat** karena model menggabungkan penilaian dari banyak pohon keputusan yang berbeda.

**Kelebihan dan Kekurangan RandomForest:**
- Kelebihan :
    + Kemampuan Generalisasi yang Baik: Model ini cenderung memberikan performa prediksi yang lebih baik pada data yang belum pernah dilihat.
    + Fleksibilitas dan Kekuatan: Random Forest dapat menangani dataset yang besar dan kompleks, menjadikannya pilihan yang cocok untuk banyak tugas pembelajaran mesin.
- Kekurangan :
    + Waktu Komputasi: Random Forest bisa menjadi lambat pada dataset yang sangat besar, tetapi ini dapat diatasi dengan menggunakan komputasi paralel (n_jobs=-1 pada GridSearchCV).

Sebagian besar titik berada dekat dengan garis diagonal, yang menunjukkan bahwa model Random Forest memiliki tingkat akurasi prediksi yang cukup baik. Nilai CLV aktual dan prediksi sebagian besar sejalan, terutama di rentang nilai CLV yang lebih rendah. 

Secara keseluruhan, terdapat tren positif yang kuat antara CLV aktual dan prediksi, menunjukkan bahwa model Random Forest mampu menangkap pola hubungan antara variabel input dan target dengan baik.
